Emission factors in openIO-Canada come in two different formats:
- basic price
- purchaser price

So in the end, if the price data you want to use the emission factors with is a purchaser price data (what you can directly read on the bill) then use the purchaser price emission factors. <br> If somehow you have access to basic prices and details on the margins and taxes, then use the basic price emission factors for the purchased commodity and the basic price emission factors for the corresponding  transportation and trade. This will lead to a more accurate assessment.

```python
# to get the basic price emission factors
tables.get_emission_factors_basic_price()
```

Note that emission factors at purchaser price are only available for commodities bought in Canada, as only the data from StatCan include details on the different margins, unlike exiobase.

```python
# to get the purchaser price emission factors
tables.get_emission_factors_purchaser_price()
```
