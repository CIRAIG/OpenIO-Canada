This module shows how to generate Environmentally Extended Input-Output (EEIO) symmetric tables for Canada using Statistiques Canada economic data and various physical flows accounts as well as data from the NPRI.

```python
import openIO
import pandas as pd
import numpy as np
```

To create an object, you simply need to enter 3 parameters:
- the path to the STATCAN economic data for the year and classification level you chose (available here: https://www150.statcan.gc.ca/n1/en/catalogue/15-602-X)
- the classification to be used for the symmetric table: either "industry" or "product"

```python
tables = openIO.IOTables('C://Users/.../Detail level/',
                         'product')
```

You now have access to typical IOTs:

Technology matrix

```python
tables.A
```

Environmental flow matrix

```python
tables.F
```

Characterization matrix

```python
tables.C
```

Final demand consumption matrix

```python
tables.Y
```

Direct emissions from final demand

```python
tables.FY
```

The openIO python module also includes a calculation function to determine emissions/impacts over the life cycle:

```python
tables.calc()
```

which then determines the Leontief inverse:

```python
tables.L
```

the total emissions linked to the final demand consumption (indirect + direct)

```python
tables.E
```

and its conversion to impacts using the IMPACT World+ impact methodology

```python
tables.D
```
