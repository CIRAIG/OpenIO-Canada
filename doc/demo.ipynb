### Contributions analysis

For the contribution analysis we require the Leontief inverse which can calculated within the .calc() function.

```python
# Leontief inverse
tables.L
```

Say we want to calculate the Quebec household consumption footprint

```python
sector = ('QC','Household final consumption expenditure')
```

We first need to diagonalize the demand

```python
diag_Y_households = pd.DataFrame(np.diag(tables.Y.loc[:, sector]),
                                                index=tables.Y.index, columns=tables.Y.index)
```

Then we calculate the contribution analysis for local consumption (within Canada)

```python
contribution_analysis_households = tables.C.dot(tables.S).dot(tables.L).dot(diag_Y_households).groupby(axis=1,level=1).sum()
```

We add to it the impact of imported goods, purchased by the intermediate sectors (i.e., NOT by final demands)

```python
contribution_analysis_households += tables.C_INT.dot(tables.SL_INT).dot(tables.IMP_matrix).dot(diag_Y_households).groupby(axis=1, level=1).sum()
```

We then add the impacts of imported good by the final demand sectors

```python
contribution_analysis_households += tables.C_INT.dot(tables.SL_INT).dot(pd.DataFrame(np.diag(tables.Y_IMP.loc[:,sector]), tables.Y_IMP.index, tables.Y_IMP.index)).droplevel(level=0,axis=1)
```

And finally we add the direct emissions of households (we dump it in a new column added to our contribution analysis matrix)

```python
contribution_analysis_households.loc[:, 'Direct emissions households'] = tables.C.dot(tables.FY).loc[:,sector]
```
